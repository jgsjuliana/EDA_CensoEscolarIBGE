```python
#Calculando a média de matrículas do ensino fundamental
#('1º ano', '2º ano', '3º ano', '4º ano', '5º ano', '6º ano', '7º ano', '8º ano', '9º ano')
#Para substituir o ano de 2021 pela média de matrículas


#Média para cada ano, desconsiderando o ano de 2021
media_ano = df.loc[df.index, ['1º ano', '2º ano', '3º ano', '4º ano', '5º ano', '6º ano', '7º ano', '8º ano', '9º ano']].mean()

media_ano
```

```python
#Média geral do ensino fundamental, para preenchimento dos valores zerados em 2021
media_fundamental = media_ano.mean()

media_fundamental
```
